```python
#This is a basic implementation to a friendly problem (housing regression problem).
import tensorflow as tf
import numpy as np
```

```python
def house_model():    
    xs = np.array([1,2,3,4,5,6], dtype = float) #Number of rooms
    ys = np.array([1,1.5,2,2.5,3,3.5], dtype = float) #Prices in hundreds of thousands
    # We build our model with 1 dense layer and 1 unit
    model = tf.keras.Sequential([tf.keras.layers.Dense(units=1,input_shape=[1])])
    # We compile your model with it optimizer and loss
    model.compile(optimizer="sgd", loss="mean_squared_error")
    # We train our model for 1000 epochss
    model.fit(xs, ys, epochs=1000)
    return model
```

```python
# We save the model
model = house_model()
```

```python
new_y = 7.0
prediction = model.predict([new_y])[0]
print(prediction)
```
